```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def vbpexam_rhs(y, x):
    return [y[1], -(x**2 - np.sin(x))*y[1] + np.cos(x)**2*y[0] + 5]
```

```python
xspan = [0, 1]
A = -3
da = 0.5

for j in range(100):
    tol = 1e-6
    y0 = [3, A]
    x = np.linspace(xspan[0], xspan[1], 100)
    ysol = odeint(vbpexam_rhs, y0, x)
    if abs(ysol[-1, 1]-5) < tol:
        break
    if ysol[-1, 1] < 5:
        A += da
    else:
        A -= da
        da /= 2

plt.plot(x, ysol[:, 0])
```
